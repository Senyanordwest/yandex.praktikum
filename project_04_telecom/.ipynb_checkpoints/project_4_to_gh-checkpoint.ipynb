# <b>Определение перспективного тарифа для телеком-компании<b>

 <font size=2>
<b>Описание проекта</b><br/>    
    
Клиентам предлагают два тарифных плана: «Смарт» и «Ультра». Чтобы скорректировать рекламный бюджет, коммерческий департамент хочет понять, какой тариф приносит больше денег.<br /> Необходимо провести предварительный анализ тарифов на небольшой выборке клиентов. В нашем распоряжении данные 500 пользователей оператора сотовой связи: кто они, откуда, каким тарифом пользуются, сколько звонков и сообщений каждый отправил за 2018 год. Нужно проанализировать поведение клиентов и сделать вывод — какой тариф лучше.

 <font size=2>
<b>Описание тарифов</b>
    

|       |<font color='limegreen'><b>Тариф «Смарт»<b></font>|<font color='tomato'><b>Тариф «Ультра»<b></font>|
| ------------- |:------------------| :-----|
| 1     | Ежемесячная плата: 550 рублей    | Ежемесячная плата: 1950 рублей |
| 2     | Включено 500 минут разговора, 50 сообщений и 15 Гб интернет-трафика |   Включено 3000 минут разговора, 1000 сообщений и 30 Гб интернет-трафика |
| 3  | Стоимость услуг сверх тарифного пакета: 1. минута разговора: 3 рубля  2. сообщение: 3 рубля; 3. 1 Гб интернет-трафика: 200 рублей.         |    Стоимость услуг сверх тарифного пакета: 1. минута разговора: 1 рубль; 2. сообщение: 1 рубль; 3. 1 Гб интернет-трафика: 150 рублей. |
    
<br />

<i>Примечание:<i><i>
 
<i>Оператор всегда округляет секунды до минут, а мегабайты — до гигабайт. Каждый звонок округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.<br />
Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.<i>

 <font size=2>
<b>Описание данных</b><br /> 


<b>Таблица users (информация о пользователях):</b> 
- `user_id` — уникальный идентификатор пользователя
- `first_name` — имя пользователя
- `last_name` — фамилия пользователя
- `age` — возраст пользователя (годы)
- `reg_date` — дата подключения тарифа (день, месяц, год)
- `churn_date` — дата прекращения пользования тарифом<br/>(если значение пропущено, то тариф ещё действовал на момент выгрузки данных)
- `city` — город проживания пользователя
- `tarif` — название тарифного плана<br /><br />

<b>Таблица calls (информация о звонках):</b>
- `id` — уникальный номер звонка
- `call_date` — дата звонка
- `duration` — длительность звонка в минутах
- `user_id` — идентификатор пользователя, сделавшего звонок<br /><br />
    
<b>Таблица messages (информация о сообщениях):</b>
- `id` — уникальный номер сообщения
- `message_date` — дата сообщения
- `user_id` — идентификатор пользователя, отправившего сообщение<br /><br />
    
<b>Таблица internet (информация об интернет-сессиях):</b>
- `id` — уникальный номер сессии
- `mb_used` — объём потраченного за сессию интернет-трафика (в мегабайтах)
- `session_date` — дата интернет-сессии
- `user_id` — идентификатор пользователя<br /><br />
    
<b>Таблица tariffs (информация о тарифах):</b>
- `tariff_name` — название тарифа
- `rub_monthly_fee` — ежемесячная абонентская плата в рублях
- `minutes_included` — количество минут разговора в месяц, включённых в абонентскую плату
- `messages_included` — количество сообщений в месяц, включённых в абонентскую плату
- `mb_per_month_included` — объём интернет-трафика, включённого в абонентскую плату (в мегабайтах)
- `rub_per_minute` — стоимость минуты разговора сверх тарифного пакета<br/>(например, если в тарифе 100 минут разговора в месяц, то со 101 минуты будет взиматься плата)
- `rub_per_message` — стоимость отправки сообщения сверх тарифного пакета
- `rub_per_gb` — стоимость дополнительного гигабайта интернет-трафика сверх тарифного пакета<br/>(1 гигабайт = 1024 мегабайта)

# Добавление и объединение данных

## Добавление в исходные таблицы данных о месяце

<font size=2>
По условиям задания в дальнейшем нам необходимо будет рассчитать показатели для каждого пользователя по месяцам.

Добавим в исходные таблицы столбец с порядковым номером месяца. Для таблицы *users* значения месяца добавим в зависимости от даты подключения тарифа, для таблиц *calls, messages* и *internet* в зависимости от даты использования услуги.

```python
users['month'] = pd.DatetimeIndex(users['reg_date']).month
users.head(10)
```

```python
calls['month'] = calls['call_date'].astype('datetime64[M]')
calls.head(10)
```

```python
messages['month'] = messages['message_date'].astype('datetime64[M]')
messages.head(10)
```

```python
internet['month'] = internet['session_date'].astype('datetime64[M]')
internet.head(10)
```

## Добавление показателей по пользователям

<font size=2> Прежде чем начать расчёт показателей, объединим данные, которые будут использоваться для расчётов и анализа в одну общую таблицу. 
Сначала создадим на основе исходных таблиц таблицы с нужной информацией по звонкам, количеству использованных сообщений и трафика в зависимости от месяца.

### Количество сделанных звонков и израсходованных минут разговора

```python
"""звонки"""
month_calls = calls.groupby(['user_id', 'month'])[
    'duration'].agg(['count', 'sum'])
month_calls.head(10)
```

### Количество отправленных сообщений

```python
"""сообщения"""
month_messages = messages.groupby(['user_id', 'month'])['id'].count()
month_messages.head(10)
```

### Объем израсходованного интернет-трафика

```python
"""интернет"""
month_internet = internet.groupby(['user_id', 'month']).sum()['mb_used']
month_internet.head(10)
```

## Добавление данных о тарифе и городе

<font size=2> Поскольку в дальнейшем нам необходимо будет проанализировать и сравнить тарифы между собой и сравнить данные по регионам, то подробные данные о клиентах - возраст, ФИО, даты подключения и отключения тарифа  нам не понадобятся, поэтому после объединения таблиц с данными о звонках, сообщениях и трафике добавим только столбцы с городом и тарифом.

```python
sum_data = pd.concat([month_calls, month_messages, month_internet],
                     axis=1, join='outer').reset_index()
sum_data
```

```python
# словари городов и тарифа
map_tariff = dict(pd.Series(users['tariff'].values, index=users['user_id']))
map_city = dict(pd.Series(users['city'].values, index=users['user_id']))
# добавление в таблицу
sum_data['tariff_name'] = sum_data['user_id'].map(map_tariff)
sum_data['city'] = sum_data['user_id'].map(map_city)
```

```python
sum_data.head(10)
```

 <a name="abcd"></a>

 <a name="abcd"></a>

```python
sum_data = sum_data.merge(tariffs, on=['tariff_name'], how='outer')
sum_data
```

<font size=2> Для дальнейшего удобства в работе с таблицой перегруппируем и переименуем столбцы:

## Расчёт помесячной выручки с каждого пользователя

```python
"""Функции для расчёта перерасхода лимитов звонков, смс и Гб, в зависимости от тарифа"""


def calls_over(row):
    if row['tariff_name'] == 'smart':
        result = row['calls_duration, min.'] - 500
        return np.where(result > 0, result, 0)
    else:
        result = row['calls_duration, min.'] - 3000
        return np.where(result > 0, result, 0)


def sms_over(row):
    if row['tariff_name'] == 'smart':
        result = row['sms_count'] - 50
        return np.where(result > 0, result, 0)
    else:
        result = row['sms_count'] - 1000
        return np.where(result > 0, result, 0)


def gb_over(row):
    if row['tariff_name'] == 'smart':
        result = row['gb_used'] - 15
        return np.where(result > 0, result, 0)
    else:
        result = row['gb_used'] - 30
        return np.where(result > 0, result, 0)


def rub_monthly_over(row):
    sum_per_month = 0
    if row['tariff_name'] == 'smart':
        sum_per_month += np.ceil(row['calls_over, min.']) * 3
        sum_per_month += row['sms_over'] * 3
        sum_per_month += np.ceil(row['gb_over']) * 200
        return sum_per_month

    else:
        sum_per_month += np.ceil(row['calls_over, min.']) * 1
        sum_per_month += row['sms_over'] * 1
        sum_per_month += np.ceil(row['gb_over']) * 150
        return sum_per_month
```

```python
sum_data['calls_over, min.'] = sum_data.apply(calls_over, axis=1)
sum_data['sms_over'] = sum_data.apply(sms_over, axis=1)
sum_data['gb_over'] = sum_data.apply(gb_over, axis=1)
sum_data['rub_monthly_over'] = sum_data.apply(rub_monthly_over, axis=1)
```

```python
(sum_data.sample(5).style.set_properties(**{'text-align': 'center'}))
```

<font size=2> Для подсчёта итоговой ежемесячной выручки с сложим значения *абонентской платы* с рассчитанными значениями *переплат*:

```python
sum_data['total_revenue'] = sum_data['rub_monthly_fee'] + \
    sum_data['rub_monthly_over']
```

# Проверка гипотез

<a name="к6."></a> <div class="alert alert-info"> <b>Перед проверкой гипотез проведём сравнение выборок с помощью тестов Левена и Бартлетта:</b><br/>
<font size="2"><i>([к списку замечаний](#К.))</i></font>

```python
"""тест Левена для выборок разделеных по тарифам"""
p = st.levene(smart_users['total_revenue'],
              ultra_users['total_revenue'], center='median')
p
```

 p-критерий стремится к нулю, нулевую гипотезу теста - одинаковость дисперсии между нашими выборками, отвергаем.

## Cредняя выручка пользователей тарифов «Ультра» и «Смарт» различаются

<font size=2> **H0 (нулевая гипотеза)** - предположение о том, что средние выручки равны.<br/>
**H1 (альтернативная гипотеза)** - средняя выручка отличается.<br/>
**α=5%**
*equal_var=False, т.к. дисперсии выборок отличаются.*

```python
alpha = .05
results = st.ttest_ind(
    smart_users['total_revenue'], ultra_users['total_revenue'], equal_var=False)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось опровергнуть нулевую гипотезу")
```

<font size=2> Нулевую гипотезу отвергаем, выручка пользователей тарифа **"Смарт"** отличается от тарифа **"Ультра"**.<br/> Посмотрим, у какого же тарифа выручка больше:

```python
print('Средняя выручка по пользователям тарифа "Смарт" - {:.2f} руб.'.format(
    smart_users['total_revenue'].mean()))
print('Средняя выручка по пользователям тарифа "Ультра" - {:.2f} руб.'.format(
    ultra_users['total_revenue'].mean()))
```

<font size=2> Несмотря на регулярные переплаты, пользователи тарифа **"Смарт"** приносят оператору меньше денег, чем пользователи тарифа **"Ультра"**. Безусловно, первые регулярно воходят за лимиты тарифа и переплачивают, но всё равно они не могут догнать вторых, у которых ежемесячная абонплата в 3,5 раза выше.<br/> 


## Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов

<font size=2> **H0 (нулевая гипотеза)** - предположение о том, что средние выручки равны.<br/>
**H1 (альтернативная гипотеза)** - средняя выручка отличается.<br/>

<font size=2> Сделаем выборки с нужнами нам данными: клиентами из Москвы и из остальных регионов:

```python
moscow_user = sum_data[sum_data['city'] == 'Москва']
other_user = sum_data[sum_data['city'] != 'Москва']
```

```python
"""тест Бартлетта для выборок разделённых по регионам"""
p = st.bartlett(moscow_user['total_revenue'], other_user['total_revenue'])
p
```

Тест Бартлетта проверяет, поступают ли образцы из выборок с равной дисперсией. Значение p-критерия < 0.05, гипотезу о равенстве дисперсий отвергаем.

```python
# moscow_user.pivot_table(
#     index=['tariff_name'], values=['calls_duration, min.', 'sms_count', 'mb_used', 'total_revenue'],
#     aggfunc={'calls_duration, min.': [np.median, np.var, np.std],
#              'sms_count': [np.median, np.var, np.std],
#              'mb_used': [np.median, np.var, np.std],
#             'total_revenue': [np.mean, np.median]})
```

```python
alpha = .05
results = st.ttest_ind(moscow_user['total_revenue'],
                       other_user['total_revenue'], equal_var=False)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось опровергнуть нулевую гипотезу")
```

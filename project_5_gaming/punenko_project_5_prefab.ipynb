```python
games.query(
    'critic_score.isnull() & rating.isnull() & user_score.isnull()').sample(3)
```

<font size=2>
Никакой зависимости пропусков от года выхода игры, жанра, платформы или продаж установить не удалось, пропуски вариативны. Идём по порядку, и рассматриваем пропуски в  данных в порядке их уменьшения:

### столбец critic_score
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
games['critic_score'].unique()
```

<font size=2>
Поскольку пропусков очень много, заменять их на среднее или медианное значение, даже с какой-либо группировкой будет некорректно, в игровой индустрии взлёты и падения серий игр нередки, как и нередки случаи высоких оценок критиков при низкой оценке игроков, каждое значение уникально, поэтому, чтобы не искажать данные заполним пропуски характерной и узнаваемой заглушкой -  <i>"-1"</i>, поскольку в исходных данных не встречается игр с отрицательным рейтингом.

```python
games['critic_score'] = games['critic_score'].fillna(-1)
```

### столбец *user_score*
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
games['user_score'].unique()
```

<font size=2>Помимо неопределённых значений, как и в столбце <b>critic_score</b> есть значения <i>'tbd'</i>. Проведя поиск узнаём значение аббревиатуры - <i>To Be Determined</i>, согласно [FAQ Metacritic](https://www.metacritic.com/faq#item13) такое значение  получает какое-либо произведение (фильм, игра, телешоу, альбом) не имеющие как минимум 4 оценок.<br/>
Данные у нас представлены практически по 17 тысячам игр начиная в период с 1980 по 2016 год, можем предположить, что отсутствие оценок может быть связано как и со старостью игры, её узконаправленным уникальном жанром, популярностью в каком-то отдельном небольшом регионе, фактически в нашем случае <i>'tbd'</i> и означает <i>'nan'</i>, поэтому произведём замену затем также как и в <b>critic_score</b> поменям пропущенные значения на заглушку и не забудем перевести столбец в числовой формат: 

```python
games['user_score'] = games['user_score'].replace(
    'tbd', np.NaN).fillna(-1).astype('float')
```

### столбец *rating*
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
games['rating'].unique()
```

<font size=2>
Ситуация скорее всего аналогичная столбцу <b>critic_score</b>. Поиск на сайте ESRB ясности не вносит, игры ищутся только по названию, без указания года выхода, процесс достоверного установления значения рейтинга, даже если игра и есть в ESRB будет довольно трудоёмким, возрастной рейтниг у игр также не всегда довольно очевиден, поэтому пропущенные значения заменим на  <i>'no rating'</i>

```python
games['rating'] = games['rating'].fillna('no rating')
```

### столбцы *year_of_release, name, genre*
<font size="2"><i>([к началу проекта](#К.))</i></font>

<font size=2>
Пускай пропуски в данных столбцах незначительны, но их неверное заполнение может исказить дальнейшие результаты, поэтому ограничимся их заменой на пустые строки = <b>" "</b>, для столбцов с названием и жанром игры. В столбце в годом выхода игры поменяем тип данных, пропуски оставим как есть, т.к заменить значения не представляется возможным.

```python
games.fillna({"name": " ", "genre": " "}, inplace=True)
```

```python
games['year_of_release'] = games['year_of_release'].astype('Int64')
```

## Выход игр по годам
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
ggames = games.groupby("year_of_release")["total_sales"].sum().reset_index()
ggames
```

```python
#sns.relplot(x="year_of_release", y="total_sales", kind="line", data=ggames, figsize=(20,10))
sns.histplot(data=ggames)

```

```python
# посчитаем число вышедших игр и проданных копий по годам
years_games = games.groupby('year_of_release')[['total_sales']].sum().join(
    games.groupby('year_of_release')[['name']].count())
# число проданных копий
trace0 = go.Scatter(x=years_games.index,
                    y=years_games.total_sales, name='Продажи, млн.копий',
                    mode='lines+markers', line_color='indianred')
# число вышедших игр
trace1 = go.Scatter(x=years_games.index,
                    y=years_games.name, name='Кол-во вышедших игр',
                    mode='lines+markers', line_color='mediumseagreen')
# определяем массив данных и задаем title графика в layout
data = [trace0, trace1]
layout = {}
# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)

fig.update_layout(
    width=950,
    height=500,
    title='Видеоигры: продажи и выход по годам',
    legend_orientation="h")
iplot(fig, show_link=False)
```

В 2000 году весной в Японии, а осенью в США начались продажи самой популярной игровой консоли в истории - <b>PlayStation 2</b>, продажи преемника, консоли <b>PlayStation 3</b> начались осенью 2005 года.</br> Продажи прямого конкурента приставкам Sony от компании Microsoft -  <b>Xbox 360</b> [стартовали](https://ru.wikipedia.org/wiki/Xbox_360#Начало_продаж) в ноябре 2005 года. Продажи <b>Nintendo DS</b> и <b>Nintendo Wii</b> начались соответственно зимой [2004](https://ru.wikipedia.org/wiki/Nintendo_DS) и -[2006](https://ru.wikipedia.org/wiki/Wii#Продажи) годов соответственно.<br/>
Таким образом мы можем выделить 5 самых популярных игрофых платформ на продажи которых приходится 55%, тогда как на остальные невошедшие в Топ-5 игровые платформы будет приходится 45% от общего числа продаж. Данные возьмём в периоде начиная с 2005 года, да 2000 год, с которого началось активное увеличение и продаж, и количество выхода игр уже не кажется таким актуальным, поскольку появилось уже не одно новое поколение игроков, которым скорее всего не будут интересны игры того периода, которые уже явно не отвечают современным веяниям в части геймдизайна и графики.

```python
"""оставляем данные с 2005 года"""
games_from_2k = games[games['year_of_release'] >= 2005]
"""формируем Топ-5 платформ"""
top_5 = games_from_2k.groupby(
    'platform')['total_sales'].sum().sort_values()[-5:].index.tolist()
# определяем массив данных и задаем title графика в layout
data = []
for platform in top_5:
    data.append(
        go.Bar(x=games_from_2k[games_from_2k.platform == platform]
               .groupby("year_of_release")['total_sales'].sum().index,
               y=games_from_2k[games_from_2k.platform == platform]
               .groupby("year_of_release")['total_sales'].sum(),
               name=platform)
    )
layout = {}
# cоздаем объект Figure и визуализируем его
fig = go.Figure(data=data, layout=layout)
fig.update_layout(
    width=950,
    height=500,
    title='Продажи по годам Топ-5 платформ',
    title_x=0.5,
    margin=dict(l=10, r=10, t=50, b=10),
    legend_orientation="h")
iplot(fig, show_link=False)
```

<font size=2>
Изучив график можно заметить, что каждая из игровых платформ в отдельности "живёт" примерно 10 лет. Примерно к середине совего жизненного цикла достигая пика продаж, за которым следует спад.<br/>
На примере истории выхода серии приставок <b>PlayStation</b> мы можем заметить, что компания как раз и старается представлять каждое следующее поколение приставок с интервалом в 5-6 лет, что в общем-то и логично, ведь ведущие производители аппаратных компонентов регулярно и ежегодно представляют и выводят на рынок новые поколения "железа". И даже если конечному потребителю - простому пользователю с его точки зрения прирост производительности будет не так заметен, исключение тут составит категория заядлых геймеров, то вот сам факт обладания приставкой десятилетней давности уже чисто психологически будет настраивать на приобретение нового "железа", особенно если уже вышло новое поколение любимой игровой платформы. ## Самые популярные платформы  (топ-5)
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
# Топ-5 по регионам
def top_region(groupby, region_sales, ax):
    pivot = games.groupby(groupby).agg({region_sales: 'sum'}).sort_values(
        by=region_sales, ascending=False)[:5]
    title_dict = {'na_sales': 'Северная Америка',
                  'eu_sales': 'Европа', 'jp_sales': 'Япония'}
    color_dict = {'na_sales': 'indianred',
                  'eu_sales': 'steelblue', 'jp_sales': 'lightpink'}
    plot = pivot.plot(kind='bar', ec='k', title=title_dict[region_sales], ax=axes[ax],
                      fontsize=18, color=color_dict[region_sales], rot=0)
    plot.legend(prop={'size': 17})
    plot.set_xlabel('')
    plot.title.set_size(20)
```

```python
fig, axes = plt.subplots(3, 1, figsize=(17, 15))
plt.suptitle("Популярные платформы", size=20)

top_region('platform', 'na_sales', 0)
top_region('platform', 'eu_sales', 1)
top_region('platform', 'jp_sales', 2)
```

<font size=2>Среди самых популярных платформ в США большей популярностью пользовались Xbox 360 и PlayStation 2, можно сказать, что они делят между собой 1 и 2 места, Wii на 3 месте по полулярности, а PlayStation 3 и Nintendo DS замыкаю Топ-5, на 4 и 5 местах соответственно.<br/> Ситуация в Европе схожая, только PlayStation 2 здесь на первом месте, PlayStation 3 на втором, а тройку лидеров замыкает Xbox 360, Wii совсем немного отстаёт от неё, а замыкает пятёрку по количеству продаж самая первая версия PlayStation.<br/> Совсем другая ситуация в Японии, здесь не нашлось места компании Microsoft, весь Топ-5 состоит из приставок "отечественного производителя". В лидерах Nintendo DS, 2-3 место делят между собой два первых поколения приставок PlayStation. На четвёртом месте раритет Super Nintendo, что было бы удивительно для остального мира, но не для Японии. И замыкает пятёрку Nintendo 3DS.

# Проверка гипотез
<font size="2"><i>([к началу проекта](#К.))</i></font>

## Средние пользовательские рейтинги платформ Xbox One и PC одинаковые
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
"""Создадит датасеты только с нужными нам для проверки гипотезы данными"""
x_one_games = (games.query('platform=="XOne"').pivot_table(
    index='name', values=('user_score'))).reset_index()
pc_games = (games.query('platform=="PC"').pivot_table(
    index='name', values=('user_score'))).reset_index()

"""убираем заглушки ("-1"), чтобы не искажать результаты и построить нормальные графики"""
x_one_games = x_one_games[x_one_games['user_score'] > 0]
pc_games = pc_games[pc_games['user_score'] > 0]
```

### Тест Левена
<font size="2"><i>([к началу проекта](#К.))</i></font>
<br/> <font size=2> Перед проверкой гипотезы посмотрим на дисперсии пользовательских оценок и сравним наши выборки с помощью теста Левена:

```python
print('Дисперсия пользовательских оценок на Xbox One составляет: {:,.2f}'
      .format(x_one_games['user_score'].var()))
print('Дисперсия пользовательских оценок на PC составляет: {:,.2f}'
      .format(pc_games['user_score'].var()))
print('Дисперсии отличаются на: {:,.2%}'
      .format((pc_games['user_score'].var()/x_one_games['user_score'].var() - 1)))
```

```python
alpha = .05
p_value = st.levene(x_one_games['user_score'], pc_games['user_score'])
p_value
```

<font size=2> Значение p-критерия достаточно велико, нулевую гипотезу теста Левена - одинаковость дисперсии между нашими выборками, можно считать подтверждённой.

### Проверка гипотезы
<font size="2"><i>([к началу проекта](#К.))</i></font>
<br/> <font size=2> <b>H0 (нулевая гипотеза)</b> - предположение о том, что средние оценки пользователей равны.<br/>
<b>H1 (альтернативная гипотеза)</b> - средние оценки отличаются.<br/>
<b>α=5%</b>
<b>equal_var=True</b>, т.к. дисперсии выборок одинаковы.

```python
alpha = .05
results = st.ttest_ind(
    x_one_games['user_score'], pc_games['user_score'], equal_var=True)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось опровергнуть нулевую гипотезу")
```

<font size=2> Значение p-критерия стремиться к нулю, можно отклонить нулевую гипотезу и утверждать, что средние пользовательские оценки платформ <b>Xbox One</b> и <b>PC</b> отличаются между собой. Посмотрим на оценки более подробно:

```python
fig = go.Figure()
fig.add_trace(go.Box(x=x_one_games['user_score'],
                     fillcolor='palegreen', line=dict(color='olivedrab'),
                     notched=True, boxmean=True, name='Xbox One', boxpoints='all'))
fig.add_trace(go.Box(x=pc_games['user_score'],
                     fillcolor='sandybrown', line=dict(color='chocolate'),
                     notched=True, boxmean=True, name='PC', boxpoints='all'))
# fig.update_traces(quartilemethod="inclusive")
fig.update_layout(
    width=950,
    height=700,
    title="Диаграммы размаха пользовательских оценок платформ Xbox One и PC",
    title_x=0.5,
    yaxis_title="Платформа",
    legend=dict(x=.5, xanchor="center", orientation="h", font_size=15),
    barmode='overlay',
    margin=dict(l=10, r=20, t=50, b=10))

fig.show()
```

<font size=2> Пользователи охотнее ставят оценки играм на <b>PC</b> и эти оценки тяготеют к более высоким значениям - средняя пользовательская оценка почти <b>7,1 балла</b>, и это хорошая, положительной оценка игры.<br/>
    Игры на <b>Xbox One</b> не могут похвастать таким же разнообразием и количеством оценок, средняя оценка даже не дотягивает до 7 баллов, и составляет <b>6,5 баллов</b>, безусловно это ещё не такая низкая и критичная оценка, но всё же это уже не твёрдые <i>7 из 10</i>. Действительно, средняя оценка игр на платформах отличается.<br/>
    И если верить пользователям, то в целом на <b>PC</b> качество игр в выше, чем на <b>Xbox One</b>.<br/>
    <i>Большее количество оценок можно объяснить не только, возможно, большим разнообразием игр на PC, но и банально удобством выставления оценки: приставки в первую очередь заточены под игры и пользователям необходимо провести большее число манипуляций, чтобы поставить оценку игре и/или написать отзыв, в отличие от PC, где если игра не понравилась, можно довольно быстро зайти в интернет и поставить оценку. </i>

## Средние пользовательские рейтинги жанров Action и Sports разные
<font size="2"><i>([к началу проекта](#К.))</i></font>

```python
"""Создадим датасеты только с нужными нам для проверки гипотезы данными"""
action_games = (games.query('genre=="Action"').pivot_table(
    index='name', values=('user_score'))).reset_index()
sport_games = (games.query('genre=="Sports"').pivot_table(
    index='name', values=('user_score'))).reset_index()

"""убираем заглушки ("-1"), чтобы не искажать результаты и построить нормальные графики"""
action_games = action_games[action_games['user_score'] > 0]
sport_games = sport_games[sport_games['user_score'] > 0]
```

### Тест Левена
<font size="2"><i>([к началу проекта](#К.))</i></font>
<br/> <font size=2> Перед проверкой гипотезы посмотрим на дисперсии пользовательских оценок и сравним наши выборки с помощью теста Левена:

```python
print('Дисперсия пользовательских оценок жанра "Action" составляет: {:,.2f}'
      .format(action_games['user_score'].var()))
print('Дисперсия пользовательских оценок жанра "Sports" составляет: {:,.2f}'
      .format(sport_games['user_score'].var()))
print('Дисперсии отличаются на: {:,.2%}'
      .format((sport_games['user_score'].var()/action_games['user_score'].var() - 1)))
```

```python
alpha = .05
p_value = st.levene(action_games['user_score'], sport_games['user_score'])
p_value
```

<font size=2> Значение p-критерия достаточно мало, чтобы мы могли утверждать одинаковость дисперсии между нашими выборками, будем считать их отличающимися.

### Проверка гипотезы
<font size="2"><i>([к началу проекта](#К.))</i></font>

<br/> <font size=2> <b>H0 (нулевая гипотеза)</b> - предположение о том, что средние оценки пользователей разные.<br/>
<b>H1 (альтернативная гипотеза)</b> - средние оценки одинаковы.<br/>
<b>α=5%</b>
<b>equal_var=False</b>, т.к. дисперсии выборок одинаковы.

```python
alpha = .05
results = st.ttest_ind(
    action_games['user_score'], sport_games['user_score'], equal_var=False)
print('P-value:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось опровергнуть нулевую гипотезу")
```

<font size=2> Значение p-критерия не настолько мало, как в случае с проверкой гипотезы по платформам, но всё же оно мало достаточно, чтобы отклонить нулевую гипотезу и утверждать, что средние пользовательские оценки жанров <b>Action</b> и <b>Sports</b> равны. Посмотрим на оценки более подробно:

```python
fig = go.Figure()
fig.add_trace(go.Box(x=action_games['user_score'],
                     fillcolor='cornflowerblue', line=dict(color='royalblue'),
                     notched=True, boxmean=True, name='Action', boxpoints='all'))
fig.add_trace(go.Box(x=sport_games['user_score'],
                     fillcolor='tan', line=dict(color='olive'),
                     notched=True, boxmean=True, name='Sports', boxpoints='all'))
# fig.update_traces(quartilemethod="inclusive")
fig.update_layout(
    width=950,
    height=700,
    title="Диаграммы размаха пользовательских оценок жанров Action и Sports",
    title_x=0.5,
    yaxis_title="Жанр",
    legend=dict(x=.5, xanchor="center", orientation="h", font_size=15),
    barmode='overlay',
    margin=dict(l=10, r=20, t=50, b=10))

fig.show()
```

<font size=2> Да, для игр жанра <b>Sports</b> средняя оценка немногим больше 6 баллов, а вот у жанра <b>Action</b> немного не дотягивает до 6,5 баллов. В целом можно сказать, что да, оценки примерно равны. Также как и при сравнении оценок по платформам, в распределении оценок по жанрам мы можем видеть, что у жанра <b>Action</b> гораздо больше пользовательских оценок, по сравнению с жанром <b>Sports</b>. Это хорошо было заметно ещё на этапе исследовательского анализа данных - игр в жанре <b>Action</b> больше, больше и продаж в этом жанре и сами игры этого жанра популярнее <b>Sports</b> у пользователей во всех регионах.

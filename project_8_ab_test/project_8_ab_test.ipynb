# Описание данных

<font size=2,5>
    
В файле <b>hypothesis.csv</b> 9 гипотез по увеличению выручки интернет-магазина с указанными параметрами `Reach`, `Impact`, `Confidence`, `Effort`.
<br/>
Структура файла:
    
* `Hypothesis` — краткое описание гипотезы;
* `Reach` — охват пользователей по 10-балльной шкале;
* `Impact` — влияние на пользователей по 10-балльной шкале;
* `Confidence` — уверенность в гипотезе по 10-балльной шкале;
* `Efforts` — затраты ресурсов на проверку гипотезы по 10-балльной шкале. <i>Чем больше значение Efforts, тем дороже проверка гипотезы.</i><br/>

Файлы <b>orders.csv</b> и <b>visitors.csv</b> содержат информацию о заказах и пользователях соответственно. Структура файлов:
    <br/>
    
<b>orders.csv</b>    
* `transactionId` — идентификатор заказа;
* `visitorId` — идентификатор пользователя, совершившего заказ;
* `date` — дата, когда был совершён заказ;
* `revenue` — выручка заказа;
* `group` — группа A/B-теста, в которую попал заказ.
    
<b>visitors.csv</b>
* `date` — дата;
* `group` — группа A/B-теста;
* `visitors` — количество пользователей в указанную дату в указанной группе A/B-теста.

<i>Сравним размеры групп в AB-тесте:

```python
visits_sum_a = visitors[visitors['group']=='A']['visitors'].sum()
visits_sum_b = visitors[visitors['group']=='B']['visitors'].sum()

print('Кол-во посетителей в группе А: {}\n'
      'Кол-во посетителей в группе B: {}'.format(visits_sum_a,
                                                 visits_sum_b)
     )
```

```python
# "сырые" данные
orders_sum_a = orders[orders['group']=='A']['transaction_id'].count()
orders_sum_b = orders[orders['group']=='B']['transaction_id'].count()
print('"Сырые" данные:')
print('Кол-во заказов в группе А: {}\n'
      'Кол-во заказов в группе B: {}'.format(orders_sum_a,
                                             orders_sum_b))
# "чистые" данные
orders_sum_a_clean = orders_clean[orders_clean['group']=='A']['transaction_id'].count()
orders_sum_b_clean = orders_clean[orders_clean['group']=='B']['transaction_id'].count()
print('-'*10*3)
print('"Чистые" данные:')
print('Кол-во заказов в группе А: {}\n'
      'Кол-во заказов в группе B: {}'.format(orders_sum_a_clean,
                                             orders_sum_b_clean))
```

```python
# находим отношение числа заказов к посещениям в каждой группе в отдельности

# "сырые" данные
p_a = orders_sum_a / visits_sum_a
p_b = orders_sum_b / visits_sum_b
print('"Сырые" данные:')
print('Конверсия в группе А: {}\n'
      'Конверсия в группе B: {}'.format(p_a, p_b))

# "чистые" данные
p_a_clean = orders_sum_a_clean / visits_sum_a
p_b_clean = orders_sum_b_clean / visits_sum_b
print('-'*10*3)
print('"Чистые" данные:')
print('Конверсия в группе А: {}\n'
      'Конверсия в группе B: {}'.format(p_a_clean, p_b_clean))
```

<font size=2,5>
В обоих случаях конверсия в тестовой группе выше, чем в расчётной.

```python
# расчеты на "сырых" данных

alpha = 0.05  # критический уровень статистической значимости
# комбинированное отношение
p_combined = (orders_sum_a + orders_sum_b) / (visits_sum_a + visits_sum_b)
#разница отношений
difference = p_a - p_b
# статистик в ст.отклонениях стандартного нормального распределения
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/visits_sum_a + 1/visits_sum_b))
# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1) 
p_value = (1 - distr.cdf(abs(z_value))) * 2  # ваш код
print('p-значение: ', p_value)
if p_value < alpha:# ваш код
    print('Отвергаем нулевую гипотезу: между размерами групп есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать размерами групп разными')
```

```python
# расчеты на "чистых" данных

alpha = 0.05  # критический уровень статистической значимости
# комбинированное отношение
p_combined_clean = (orders_sum_a_clean + orders_sum_b_clean) / (visits_sum_a + visits_sum_b)
#разница отношений
difference_clean = p_a_clean - p_b_clean
# статистик в ст.отклонениях стандартного нормального распределения
z_value_clean = difference_clean / mth.sqrt(p_combined * (1 - p_combined_clean) * (1/visits_sum_a + 1/visits_sum_b))
# задаем стандартное нормальное распределение (среднее 0, ст.отклонение 1)
distr = st.norm(0, 1) 
p_value_clean = (1 - distr.cdf(abs(z_value_clean))) * 2  # ваш код
print('p-значение: ', p_value_clean)
if p_value_clean < alpha:# ваш код
    print('Отвергаем нулевую гипотезу: между размерами групп есть значимая разница')
else:
    print(
        'Не получилось отвергнуть нулевую гипотезу, нет оснований считать размерами групп разными')
```

#### Подготовка данных перед построением графиков

```python
# создаём датафрейм с уникальными парами значений `date` и `group`

dates_groups = orders_clean[['date','group']].drop_duplicates()
#dates_groups
```

```python
#orders.head(1)
```

```python
# проводим аггрегацию по заказам, сохраняем в переменную

orders_aggregated = dates_groups.apply(
    lambda x: orders_clean[
        np.logical_and(
            orders_clean['date'] <= x['date'], orders_clean['group'] == x['group']
        )
    ].agg(
        {'date': 'max',
         'group': 'max',
         'transaction_id': pd.Series.nunique,
         'visitor_id': pd.Series.nunique,
         'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])

orders_aggregated.head(7)
```

```python
# проводим аггрегацию по пользователям, сохраняем в переменную

visitors_aggregated = dates_groups.apply(
    lambda x: visitors[
        np.logical_and(
            visitors['date'] <= x['date'], visitors['group'] == x['group']
        )
    ].agg(
        {'date' : 'max',
         'group' : 'max',
         'visitors' : 'sum'}),
    axis=1
).sort_values(by=['date','group'])

visitors_aggregated.head(7)
```

```python
# объединяем созданые выше переменные

cumulative_data = orders_aggregated.merge(visitors_aggregated, left_on=['date', 'group'], right_on=['date', 'group'])
#cumulative_data.head(7)
```

#### Статистическая значимость различий в конверсии между группами по «сырым» данным.

```python
#orders.head(1)
```

<font size=2,5>
Подготовим данные:

```python
# агрегированные кумулятивные по дням, посетители, группа А
visitors_a_daily = visitors[visitors['group'] == 'A'][['date', 'visitors']]
visitors_a_daily.columns = ['date', 'visitors_per_date_a']

visitors_a_cummulative = visitors_a_daily.apply(lambda x: visitors_a_daily[visitors_a_daily['date'] <= x['date']]
                                                .agg({'date': 'max', 'visitors_per_date_a': 'sum'}), axis=1,)

visitors_a_cummulative.columns = ['date', 'visitors_cummulative_a']

# агрегированные кумулятивные по дням, посетители, группа B
visitors_b_daily = visitors[visitors['group'] == 'B'][['date', 'visitors']]
visitors_b_daily.columns = ['date', 'visitors_per_date_b']

visitors_b_cummulative = visitors_b_daily.apply(lambda x: visitors_b_daily[visitors_b_daily['date'] <= x['date']]
                                                .agg({'date': 'max', 'visitors_per_date_b': 'sum'}), axis=1,)

visitors_b_cummulative.columns = ['date', 'visitors_cummulative_b']

# агрегированные по дням о заказах и выручке, группа А
orders_a_daily = (orders[orders['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False).agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'}))

orders_a_daily.columns = ['date', 'orders_per_date_a', 'revenue_per_date_a']

orders_a_cummulative = orders_a_daily.apply(
    lambda x: orders_a_daily[orders_a_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_a': 'sum', 'revenue_per_date_a': 'sum'}), axis=1,).sort_values(by=['date'])

orders_a_cummulative.columns = ['date', 'orders_cummulative_a', 'revenue_cummulative_a']

# агрегированные по дням о заказах и выручке, группа B
orders_b_daily = (orders[orders['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
                  .groupby('date', as_index=False).agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'}))

orders_b_daily.columns = ['date', 'orders_per_date_b', 'revenue_per_date_b']

orders_b_cummulative = orders_b_daily.apply(
    lambda x: orders_b_daily[orders_b_daily['date'] <= x['date']].agg(
        {'date': 'max', 'orders_per_date_b': 'sum', 'revenue_per_date_b': 'sum'}), axis=1,).sort_values(by=['date'])

orders_b_cummulative.columns = ['date', 'orders_cummulative_b', 'revenue_cummulative_b']
```

```python
# Объединяем по дате все полученные кумулятивные данные в один датафрейм
data = (
    orders_a_daily.merge(
        orders_b_daily, left_on='date', right_on='date', how='left'
    )
    .merge(orders_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(orders_b_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_a_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_b_daily, left_on='date', right_on='date', how='left')
    .merge(visitors_a_cummulative, left_on='date', right_on='date', how='left')
    .merge(visitors_b_cummulative, left_on='date', right_on='date', how='left')
)
data.head(5)
```

```python
# переменные для пользователей, сделавших заказы

orders_by_users_a = (orders[orders['group'] == 'A'].groupby('visitor_id', as_index=False)
                  .agg({'transaction_id': pd.Series.nunique}))
orders_by_users_a.columns = ['visitor_id', 'orders']
 
orders_by_users_b = (orders[orders['group'] == 'B'].groupby('visitor_id', as_index=False)
                  .agg({'transaction_id': pd.Series.nunique}))
orders_by_users_b.columns = ['visitor_id', 'orders']
```

```python
# переменные в которых пользователям с заказами будет соответствовать число заказов пользователя
# пользователям без заказов — нули

sample_a = pd.concat([orders_by_users_a['orders'],
                     pd.Series(0, index=np.arange(data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])),
                               name = 'orders')], axis = 0)
 
sample_b = pd.concat([orders_by_users_b['orders'],
                     pd.Series(0, index=np.arange(data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])),
                               name = 'orders')], axis = 0)
```

<b>Статистическая значимость различий в конверсии между группами по «сырым» данным

<font size=2,5>
    
Зададим условия для для проведения теста Манна-Уитни на исследуемых выборках:<br/>
* Пороговый уровень статистической значимости возьмём в 5%<br/>
<b>alpha = .5</b><br/>
    
<b>H0</b> - Вероятность того, что наблюдение из группы A превышает наблюдение из группы B равна 50%.<br/>
<b>H1</b> - Такая вероятность не равна 50%.<br/>

```python
print('p-value = {0:.3f}'.format(stats.mannwhitneyu(sample_a, sample_b)[1]))

print('Вероятность того, что наблюдение из группы A превышает наблюдение из группы B равна 50% составляет: {0:.2%}'
      .format((data['orders_per_date_b'].sum() / data['visitors_per_date_b'].sum())
              / (data['orders_per_date_a'].sum() / data['visitors_per_date_a'].sum())-1))
```

<font size=2,5>
    
p-value меньше 0.05, нулевую гипотезу отвергаем.<br/> Вероятность того, что наблюдение из группы A превышает наблюдение из группы B составляет: 13.81%.

#### Статистическая значимость различий в среднем чеке заказа между группами по «сырым» данным.

```python
print('p-value = {0:.3f}'.format(stats.mannwhitneyu(orders[orders['group']=='A']['revenue'],
                                                    orders[orders['group']=='B']['revenue'])[1]))

print('Относительный прирост среднего чека группы B по отношению к группе A: {0:.2%}'
      .format(orders[orders['group']=='B']['revenue'].mean()
              /orders[orders['group']=='A']['revenue'].mean() - 1))
```

<font size=2,5>
    
p-value больше 0.05, нулевую гипотезу о том, что статистически значимых различий в среднем чеке между группами нет, не опровергаем.<br/>
При этом разница в среднем чеке составляет почти 26% в пользу тестовой группы. Тут можно вспомнить, что пользователь, совершивший заказ на 1,3 млн. у.е. как раз и был в группе B, определённо тут сказывается влияние этого заказа.

#### Статистическая значимость различий в конверсии между группами по «очищенным» данным.

<font size=2,5>
Подготовим "чистые"данные:<br/>
Аномальными посчитаем пользователей соверших больше трёх заказов - 99 перцентиль в "сырых" данных - 4 заказа и пользователей с заказом на сумму больше 28 тыс., что соответствует 95 перцентилю.

```python
# пользователи с числом заказов больше 3

users_with_many_orders = pd.concat([orders_by_users_a[orders_by_users_a['orders'] > 3]['visitor_id'],
                                 orders_by_users_b[orders_by_users_b['orders'] > 3]['visitor_id'],],axis=0,)

# пользователи с выручкой больше 28 тыс.
users_with_expensive_orders = orders[orders['revenue'] > 28000]['visitor_id']

# объединяем их в таблице
abnormal_users = (pd.concat([users_with_many_orders, users_with_expensive_orders], axis=0)
                  .drop_duplicates().sort_values())
#abnormalUsers.head(5)
print('Количество "аномальных" пользователей:', abnormal_users.shape[0])
```

<font size=2,5>
Посмотрим, оказывают ли они влияние на результаты теста.

```python
# создаём переменные с чистыми данными по заказам

sample_a_filtered = pd.concat([orders_by_users_a[
    np.logical_not(orders_by_users_a['visitor_id'].isin(abnormal_users))]['orders'],
        pd.Series(0, index=np.arange(data['visitors_per_date_a'].sum() - len(orders_by_users_a['orders'])),
                  name='orders',),], axis=0,)

sample_b_filtered = pd.concat([orders_by_users_b[
    np.logical_not(orders_by_users_b['visitor_id'].isin(abnormal_users))]['orders'],
        pd.Series(0, index=np.arange(data['visitors_per_date_b'].sum() - len(orders_by_users_b['orders'])),
                  name='orders',),], axis=0,)
```

<b>Статистическая значимость различий в конверсии между группами по «очищенным» данным

```python
print('p-value = {0:.3f}'.format(stats.mannwhitneyu(sample_a_filtered, sample_b_filtered)[1]))

print('Относительный прирост конверсии группы B по отношению к группе A: {0:.2%}'
      .format((sample_b_filtered.mean()/sample_a_filtered.mean()-1)))
```

<font size=2,5>
    
p-value меньше 0.05, нулевую гипотезу о том, что статистически значимых различий в конверсии между "чистыми" группами нет, отвергаем. Разница в конверсии статистически значима, а избавившись от "аномальных" пользователей группа B теперь выигрывает в конверсии у группы А почти на 15%.

#### Статистическая значимость различий в среднем чеке заказа между группами по «очищенным» данным.

```python
print('p-value = {0:.3f}'.format(
    stats.mannwhitneyu(
        orders[np.logical_and(orders['group'] == 'A',
                              np.logical_not(orders['visitor_id'].isin(abnormal_users)),)]['revenue'],
        orders[np.logical_and(orders['group'] == 'B',
                              np.logical_not(orders['visitor_id'].isin(abnormal_users)),)]['revenue'],)[1]))

print('Относительный прирост среднего чека группы B по отношению к группе A: {0:.2%}'.format(orders[
        np.logical_and(orders['group'] == 'B', np.logical_not(orders['visitor_id']
                                                             .isin(abnormal_users)),)]['revenue'].mean()
        / orders[np.logical_and(orders['group'] == 'A', np.logical_not(orders['visitor_id']
                                                                      .isin(abnormal_users)),)]['revenue'].mean()- 1))
```

<font size=2,5>
    
p-value больше 0.05, нулевую гипотезу о том, что есть статистически значимые различия между "чистыми" средними чеками, мы опровергнуть не можем. Разницы в среднем чеке "чистых" групп нет, тестовая группа проигрывает контрольной всего 1,7%.
